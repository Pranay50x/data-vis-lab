```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Load and Pre-process Data
df = pd.read_csv("AirPassengers.csv")

# Convert to datetime and set index
df['Month'] = pd.to_datetime(df['Month'])
df.set_index('Month', inplace=True)

# 2. Calculate 12-month Moving Average
df['12-MA'] = df['Passengers'].rolling(window=12).mean()

# Trend Visualization (Actual + Moving Average)
plt.figure()
plt.plot(df['Passengers'])
plt.plot(df['12-MA'])
plt.title("Airline Passengers Trend")
plt.show()

# 3. Seasonal Patterns (Boxplot)
df['Month_Name'] = df.index.month_name()

months_order = ['January', 'February', 'March', 'April', 'May', 'June',
                'July', 'August', 'September', 'October', 'November', 'December']

plt.figure()
sns.boxplot(data=df, x='Month_Name', y='Passengers', order=months_order, palette='viridis', hue='Month_Name')
plt.title("Seasonal Patterns by Month")
plt.xticks(rotation=45)
plt.show()
```
