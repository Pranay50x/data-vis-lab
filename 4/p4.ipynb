```python
import pandas as pd
import plotly.express as px
from dash import Dash, dcc, html
from dash.dependencies import Input, Output

# Load dataset
df = pd.read_csv("tips.csv")

# Create app
app = Dash(__name__)

# Layout
app.layout = html.Div([
    html.H1("Tips Dashboard"),

    dcc.Dropdown(
        id='day',
        options=[{'label': d, 'value': d} for d in df['day'].unique()],
        value=df['day'].unique()[0]
    ),

    dcc.Graph(id='scatter'),
    dcc.Graph(id='bar')
])

# Callback
@app.callback(
    [Output('scatter', 'figure'),
     Output('bar', 'figure')],
    [Input('day', 'value')]
)
def update(day):
    dff = df[df['day'] == day]

    fig1 = px.scatter(dff, x='total_bill', y='tip', color='sex',
                      title="Bill vs Tip")

    fig2 = px.bar(dff, x='sex', y='tip', color='sex',
                  title="Tips by Gender")

    return fig1, fig2

# Run app
if __name__ == "__main__":
    app.run(debug=True)
```
